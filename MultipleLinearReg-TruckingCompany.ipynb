```python
## getting the data in tabular form
import pandas as pd
## x1 --> miles travelled
dataframe = pd.read_excel("/home/ashish/projects/MultipleLinearRegression_TruckingCompany/TRUCKING.xlsx") 
dataframe
```

# SCATER PLOT-3 => MERGING 1 AND 2

```python
import matplotlib.pyplot as plt

## Multiple regression Plotting
plt.figure()
plt.scatter(dataframe['x1'],dataframe['travel_time'],color="green")
plt.scatter(dataframe['n_of_deliveries'],dataframe['travel_time'],color="red")
plt.ylabel("Travel time")
plt.xlabel("x1(miles travelled) in green and n_of_deliveries in red")
plt.title("MULTIPLE REGRESSION")
plt.show()
```

```python
## Linear Regression Modelling ---> for travel time as dependent variable and x1(miles travelled) as the only independent variablee

from statsmodels.formula.api import ols

linear_model = ols(formula="travel_time ~ x1",data=dataframe).fit()
linear_model.summary()
```

# NOTE::
    1. So, linear regression equation ==> travel_time = 1.2739 + 0.06*x1
    2. here R-squared and adjusted R-squared are low i.e. the variability of the model cannot be efficiently explained using only x1(miles travlled)

```python
## Multiplen regression modellling

from statsmodels.formula.api import ols
multiple_model = ols(formula="travel_time ~ x1+n_of_deliveries",data=dataframe).fit()
multiple_model.summary()
```

# Note:: 
    1. Multiple linear regression equation: 
        travel_time= -0.8687 +0.061*x1+0.923*n_of_deliveries.
       
    2. High R-squared and adjusted R-squared, so variability of the model can be efficiently explained with the help of x1(miles travelled) and n_of_deliveries.
    
    3. Since, Prob (F-statistic):	0.000276  <= 0.05  this implies the model is overall significant.
    
    4. Also, since for x1 => p-value = 0.000 <= 0.05 so x1 is also a significant variable.
    
    5. Also, since for n_of_deliveries = 0.004 <= 0.05 so n_of_deliveries is also a significant variable.

```python
## ANOVA table for linear regression model
from statsmodels.stats.anova import anova_lm

anova_table_lin = anova_lm(linear_model)
anova_table_lin
```

```python
## Anova table for multiple regression model
from statsmodels.stats.anova import anova_lm

anova_table_mul = anova_lm(multiple_model,typ=1)
anova_table_mul
```

NOTE:: 
    RSM(Response Surface Model) = Multiple Regression Model
